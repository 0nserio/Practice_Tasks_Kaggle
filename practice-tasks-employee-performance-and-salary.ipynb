```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
## 1. Loading the Dataset and reviewing the first 10 rows
df = pd.read_csv("/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv")
df.head(10)
```

```python
## 2. Reviewing the columns and data types
df.dtypes
```

```python
## 3. Number of rows and columns 
df.shape
```

```python
## 4. Selecting the name, age and salary and using iloc to select the first 5 rows
dt = df[['Name', 'Age', 'Salary']]
dt.iloc[:5]
```

```python
## 5. selecting rows using loc method where the Department is "Sales"
df.loc[df['Department']=='Sales']
```

```python
## 6. rows where the Salary is greater than 8000 and Status is "Active"
df.loc[(df['Salary']> 800) | (df['Status']== 'Active')]
```

```python
## 7. rows where the employee has been with the company for more than 5 years and has a Performance Score of at least 4.
df.loc[(df['Experience']> 5) & (df['Performance Score'] >= 4)]
```

```python
## 9. every alternate row from the DataFrame using slicing with iloc
df.iloc[::2]
```

```python
## 10. rows where the employee belongs to "Sales" and has a Salary in the top 10% of all salaries
##  salary threshold of the top 10%
salary_threshold = df['Salary'].quantile(0.9)

top_sales_employees = df.loc[(df['Department']== 'Sales')&(df['Salary']>= salary_threshold)]
print(top_sales_employees)
```

```python
## 11. Replace all "Inactive" values in the Status column with "Retired"
df['Status'] = df['Status'].replace('Inactive', 'Retired')
print(df)
```

```python
## 12. new column Bonus where values are 10% of the Salary column.
df['Bonus'] = df['Salary']*0.10
df
```

```python
## 13. Update the Performance Score of employees in the "IT" department to 5 if it is NaN.
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5
df
```

```python
## 14. Modify the Experience of employees with less than 2 years to 2.
df.loc[df['Experience']<2, 'Experience'] = 2
df
```

```python
## 15. Normalize the Salary column so that all values are between 0 and 1.
df['Normalized_Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
print(df['Normalized_Salary'])
```

```python
## 16. Replace all Location values with "Remote" for employees with a Performance Score below 3.
df.loc[df['Performance Score']< 3, 'Location']= 'Remote'
df
```

```python
## 17. new column Experience Level with values "Junior", "Mid", or "Senior" based on the Experience.
df['Experience Level'] = pd.cut(df['Experience'],
                                bins=[-1, 3, 7, float('inf')],
                                labels=['Junior', 'Mid', 'Senior'],
                                right=False)
df
```

```python
## 18. Add a new row with specific values using the append() method
new_row = {'Name': 'John Doe', 'Age': 30, 'Department': 'HR', 'Salary': 6000, 'Status': 'Active', 'Years_with_Company': 3, 'Performance_Score': 4, 'Experience': 3, 'Location': 'New York'}
df = df._append(new_row, ignore_index=True)
print(df.tail())
```

```python
## 19. Delete the column Session from the DataFrame
df_ses=df.drop(columns=['Session'], inplace=False, errors='ignore')
print(df_ses.head())
```

```python
## 20: Delete rows where the Age is greater than 60
df_age60 = df[df['Age'] <= 60]
print(df_age60.head())
```

```python
## 21. Add a new column Age Group with values "Young", "Middle-aged", or "Senior" based on Age
df['Age Group'] = pd.cut(df['Age'],
                         bins=[0, 30, 50, float('inf')],
                         labels=['Young', 'Middle-aged', 'Senior'],
                         right=False)

# Display the first few rows to confirm
print(df.head())
```

```python
## 22. Identify all columns with missing values using the isna() method.
df.isna().sum()
```

```python
## 23. Fill the missing values in Performance Score with the average score of that column.
df['Performance_Score'].fillna(df['Performance_Score'].mean(), inplace=True)
print(df.head())
```

```python
## 24. Drop all rows with missing values using the dropna() method
df1= df.dropna(inplace=False)
print(df1.shape)
```

```python
## 25: Replace missing values in Performance Score with the median of the scores, grouped by Department
df['Performance_Score'] = df.groupby('Department')['Performance_Score'].transform(lambda x: x.fillna(x.median()))
print(df.head())
```

```python
## 26: Identify rows where the employee is in the "IT" department and their Age is above the average Age of "IT" employees
it_avg_age = df.loc[df['Department'] == 'IT', 'Age'].mean()
print(df.loc[(df['Department'] == 'IT') & (df['Age'] > it_avg_age)])
```

```python
## 27: Identify rows with duplicate Name and Joining Date combinations, and remove duplicates
df.drop_duplicates(subset=['Name', 'Joining Date'], inplace=True)
print(df.head())
```

```python
## 28: Remove rows where the Experience value is inconsistent with the Age (e.g., Experience > Age - 18)
df = df[df['Experience'] <= df['Age'] - 18]
print(df.head())
```

```python
## 29: Sort the DataFrame by Salary in descending order
print(df.sort_values(by='Salary', ascending=False))
```

```python
##  30: Sort the DataFrame by Department and then by Joining Date in ascending order
print(df.sort_values(by=['Department', 'Joining Date']))
```

```python
##  31: Rank all employees by Salary within their Department and add the rank as a new column
df['Salary_Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)
print(df.head())
```

```python
## 32: Sort the dataset by Experience in descending order, but break ties using Performance Score in ascending order
print(df.sort_values(by=['Experience', 'Performance_Score'], ascending=[False, True]))
```

```python
## 33: Create a new DataFrame containing only employees in "HR" with a Salary above 5000
hr_df = df[(df['Department'] == 'HR') & (df['Salary'] > 5000)]
print(hr_df)
```

```python
## 34: Filter out all rows where the Location is "Chicago"
df_Chicago = df[df['Location'] != 'Chicago']
print(df_Chicago.head())
```

```python
## 35: Delete all rows where the Performance Score is below 2
df2 = df[df['Performance_Score'] >= 2]
print(df2.head())
```

```python
## 36: Find the average Salary for each Department
print(df.groupby('Department')['Salary'].mean())
```

```python
## 37: Calculate the total Salary of all employees grouped by Status
print(df.groupby('Status')['Salary'].sum())
```

```python
## 38: Find the maximum Performance Score for employees in each Location
print(df.groupby('Location')['Performance_Score'].max())
```

```python
## 39: Calculate the total Salary paid to employees who joined in each year
df['Joining_Year'] = pd.to_datetime(df['Joining Date']).dt.year
print(df.groupby('Joining_Year')['Salary'].sum())
```

```python
## 40: Find the department with the highest average Performance Score
print(df.groupby('Department')['Performance Score'].mean().idxmax())
```

```python
## 41: Count the number of employees in each Location grouped by their Session
print(df.groupby(['Location', 'Session']).size())
```

```python
## 42: Group employees by Status and calculate the sum, mean, and standard deviation of their Salary
print(df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std']))
```

```python
## 43: Use the apply() method to create a new column Monthly Salary by dividing the Salary by 12
df['Monthly_Salary'] = df['Salary'].apply(lambda x: x / 12)
print(df.head())
```

```python
## 44: Use the applymap() method to convert all string values in the DataFrame to uppercase
string_columns = df.select_dtypes(include='object').columns
df[string_columns] = df[string_columns].applymap(lambda x: x.upper() if isinstance(x, str) else x)
print(df.head())
```

```python
## 45: Create a second dataset containing Department and Budget. Perform an inner join with the original dataset
budget_df = pd.DataFrame({
    'Department': ['IT', 'HR', 'Sales'],
    'Budget': [100000, 80000, 120000]
})
merged_df = pd.merge(df, budget_df, on='Department', how='inner')
print(merged_df.head())
```

```python
## 46: Merge the dataset with another dataset that includes additional information about employee ID and their recent Projects
projects_df = pd.DataFrame({
    'ID': df['ID'].sample(5),
    'Recent_Projects': [3, 2, 5, 1, 4]
})
merged_projects_df = pd.merge(df, projects_df, on='ID', how='inner')
print(merged_projects_df.head())
```

```python
## 47: Perform a left join with a dataset containing Department and Team Lead details, adding Team Lead to the DataFrame
team_lead_df = pd.DataFrame({
    'Department': ['IT', 'HR', 'Sales'],
    'Team_Lead': ['Alice', 'Bob', 'Charlie']
})
df = pd.merge(df, team_lead_df, on='Department', how='left')
print(df.head())
```

```python
## 48: Create a DataFrame with employee ID and Project Count, then merge it with the original dataset. Calculate the average Salary per project
project_count_df = pd.DataFrame({
    'ID': df['ID'].sample(5),
    'Project_Count': [2, 4, 6, 3, 1]
})
merged_project_count_df = pd.merge(df, project_count_df, on='ID', how='inner')
merged_project_count_df['Salary_per_Project'] = merged_project_count_df['Salary'] / merged_project_count_df['Project_Count']
print(merged_project_count_df.head())
```

```python
## 49: Calculate the percentage increase in Salary for each employee compared to the average Salary in their Department
 avg_salary_by_dept = df.groupby('Department')['Salary'].transform('mean')
df['Salary_Increase_%'] = ((df['Salary'] - avg_salary_by_dept) / avg_salary_by_dept) * 100
print(df.head())
```
